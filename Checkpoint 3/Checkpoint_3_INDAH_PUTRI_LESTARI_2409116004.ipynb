# Pola Pembelian Tiket

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["Number_of_Person"], bins=6, kde=True, color="skyblue")
plt.title("Distribusi Jumlah Tiket dalam Satu Transaksi")
plt.xlabel("Jumlah Tiket")
plt.ylabel("Frekuensi")
plt.show()
```

A. Insight

2. Ada peningkatkan tiket dalam jumlah 6 dan 7 tiket, yang kemungkinan besar berasal dari grup atau keluarga.
3. Pembelian tiket dalam jumlah 2-5 tiket relatif rendah, yang menunjukkan penonton lebih jarang membeli dalam kelompok kecil.

B. Action
1. Optimasi promosi untuk penonton tunggal, misalnya program loyalitas atau diskon setelah menonton beberapa film.

3. Membuat promosi, seperti Buy 1 Get 1 untuk kelompok kecil.

# Analisis Tren Penjualan Tiket berdasarkan Genre Film

```python
plt.figure(figsize=(8, 5))
genre = df.groupby("Movie_Genre")["Number_of_Person"].sum().reset_index()
sns.lineplot(x="Movie_Genre", y="Number_of_Person", data=genre, marker="o", color="blue", linewidth=2)
plt.title("Tren Penjualan Tiket Berdasarkan Genre Film")
plt.xlabel("Genre Film")
plt.ylabel("Jumlah Tiket Terjual")
plt.grid()
plt.show()
```

A. Insight
1. Penjualan mengalami penurunan dari Action ke Comedy, lalu sedikit meningkat di Drama dan mencapai puncaknya kembali di Horror.
2. Penurunan drastis terjadi pada genre Sci-Fi, yang menunjukkan bahwa Sci-Fi memiliki daya tarik yang jauh lebih rendah dibandingkan genre lainnya.

B. Action
1. Mempertahankan performa genre film Action dan Horror, seperti lebih banyak jadwal tayang dan promosi.
2. Karena penjualan tiket genre Drama dan Comedy relatif stabil, tetap melakukan promosi, seperti harga tiket lebih murah di jam tertentu untuk menarik lebih banyak penonton.
3. Memerlukan strategi pemasaran secara besar-besaran, seperti paket bundling, screening eksklusif, atau kerja sama dengan komunitas pecinta Sci-Fi  untuk meningkatkan ketertarikan.




# Analisis Preferensi Film

```python
plt.figure(figsize=(8, 5))
df.groupby("Movie_Genre")["Number_of_Person"].sum().reset_index()
sns.barplot(x="Movie_Genre", y="Number_of_Person",hue='Movie_Genre' , data=genre, palette="viridis")
plt.xlabel("Genre Film")
plt.ylabel("Jumlah Penonton")
plt.title("Jumlah Penonton per Genre Film")
plt.xticks(rotation=45)
plt.show()
```

A. Insight
1. Horror dan Action memiliki total penjualan tiket tertinggi mendekati atau di atas 800 penonton, yang menunjukkan daya tarik besar bagi penonton.
2. Comedy dan Drama memiliki total penjualan yang hampir sama mendekati atau sedikit di atas 800 penonton.
3. Sci-Fi memiliki total penjualan tiket paling rendah di bawah 800 penonton, mungkin dikarenakan jumlah film lebih sedikit atau kurang populer sehingga peminatnya sedikit.

B. Action
1. Mempertahankan dan meningkatkan jadwal pemutaran film Horror dan Action untuk mengakomodasi permintaan tinggi, namun tetap meningkatkan genre film lainnya untuk memberikan pengalaman penonton yang baik.
2. Membangun komunitas penggemar Sci-Fi dengan event seperti diskusi film, meet-up penggemar, serta berkolaborasi dengan brand teknologi canggih dan inovasi digital.


# Analisis Pemilihan Kursi

```python
plt.figure(figsize=(8, 5))
genre = df.groupby("Seat_Type")["Number_of_Person"].sum().reset_index()
sns.barplot(x="Seat_Type", y="Number_of_Person",hue='Seat_Type' , data=genre, palette="coolwarm")
plt.title('Jumlah Penonton per Tipe Kursi')
plt.xlabel('Tipe Kursi')
plt.ylabel('Jumlah Penonton')
plt.xticks(rotation=45)
plt.show()
```

A. Insight

*   Penonton yang lebih memilih harga terjangkau (Standard).
*   Penonton yang bersedia membayar lebih untuk kenyamanan ekstra dan pengalaman lebih mewah.

B. Action

# Korelasi Pembelian Ulang atau Kunjungan Berulang

```python
df_encoded = pd.get_dummies(df, columns=['Movie_Genre', 'Seat_Type'])
plt.figure(figsize=(14,6))
sns.heatmap(df_encoded.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Korelasi Antar Variabel")
plt.show()
```

```python
df_encoded = pd.get_dummies(df, columns=['Movie_Genre', 'Seat_Type'])

# Korelasi hanya dengan kolom Purchase_Again
korelasi_purchase_again = df_encoded.corr()["Purchase_Again"].sort_values(ascending=False)

plt.figure(figsize=(8, 5))
sns.barplot(x=korelasi_purchase_again.index, y=korelasi_purchase_again.values, hue=korelasi_purchase_again.index, palette='coolwarm')
plt.title("Korelasi Variabel dengan Pembelian Ulang")
plt.xlabel("Variabel")
plt.ylabel("Nilai Korelasi")
plt.xticks(rotation=45, ha='right')
plt.show()
```

A. Insight
1. Genre film Drama dan Horror memiliki korelasi rendah (0.047 dan 0.028), yang menunjukkan bahwa kedua genre film ini memiliki sedikit pengaruh terhadap keputusan penonton untuk melakukan pembelian ulang.
2. Tipe kursi Standar memiliki korelasi positif yang kecil (0.047), yang menunjukkan bahwa penonton yang memilih kursi ini sedikit lebih cenderung untuk melakukan pembelian ulang dibandingkan tipe kursi lainnya.
3. Age memiliki korelasi sangat rendah (0.023), yang menunjukkan bahwa usia pelanggan tidak terlalu berpengaruh terhadap pembelian ulang.
4. Genre film Action, Sci-Fi, dan Comedy memiliki korelasi yang sangat kecil dibandingkan dengan genre lainnya, yang menunjukkan bahwa ketiga genre film ini tidak memiliki pengaruh terhadap penonton untuk melakukan pembelian ulang. Namun, kecilnya korelasi ini mungkin disebabkan oleh kurangnya pengalaman atau ketertarikan penonton terhadap genre tersebut.
5. Tipe kursi Premium dan VIP memiliki korelasi negatif sangat kecil (-0.013 dan -0.034), yang menunjukkan bahwa kursi Premium dan VIP tidak mempengaruhi keputusan penonton untuk melakukan pembelian ulang. Namun, kecilnya korelasi negatif ini bisa jadi disebabkan oleh kurangnya kenyamanan atau pengalaman yang kurang memuaskan, sehingga penonton enggan untuk kembali.
6. Harga tiket dan total harga tiket memiliki korelasi negatif sangat kecil (-0.023), yang menunjukkan bahwa harga tiket dan total harga tiket yang dibayarkan tidak memiliki pengaruh terhadap keputusan untuk melakukan kunjungan berulang. Namun, harga tiket yang lebih tinggi mungkin menjadi penghalang bagi beberapa kelompok penonton.
7. Dikarenakan korelasi faktor dalam dataset rendah, artinya ada kemungkinan faktor lain, seperti layanan, kebersihan, dan kenyamanan bioskop memiliki pengaruh terhadap keputusan penonton untuk melakukan kunjungan berulang.
